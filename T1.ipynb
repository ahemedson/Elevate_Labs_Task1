```python
#importing pandas library
import pandas as pd
```

```python
#reading the dataset and converting it into a dataframe
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
df.head()
```

```python
#checking for null values
print(df.isnull())
```

```python
#count of null values if there are any
print(df.isnull().sum())
```

```python
#removing rows with missing values
df = df.dropna()
```

```python
#removing the duplicate rows
df.drop_duplicates()
```

```python
#standardizing the values of the column Gender into lowercase
df['Gender'] = df['Gender'].str.lower()
```

```python
df
```

```python
#standardizing the values of the column Gender
df['Gender'] = df['Gender'].replace({'m':'male','f':'female','Male':'male','Female':'female'})
```

```python
df
```

```python
#removing the unwanted and the trailing spaces from the values in the column Gender
df['Gender'] = df['Gender'].str.strip()
```

```python
df
```

```python
#removing the unwanted and the trailing spaces,standardizing the column headers to lowercase
df.columns = df.columns.str.strip().str.lower()
```

```python
df
```

```python
#replacing spaces with underscore
df.columns = df.columns.str.replace(' ','_')
```

```python
df
```

```python
#replacing special characters
df.columns = df.columns.str.replace(r'[^a-z0-9_]','',regex=True)
```

```python
df
```

```python
#checking the datatypes of each column
print(df.dtypes)
```

```python
#converting gender to category dtype to save memory and potentially speed up analysis
df['gender'] = df['gender'].astype('category')
```

```python
print(df.dtypes)
```

```python
#lookup table for each category
print(df['gender'].cat.categories)
```

```python
#integer codes representing each category for every row
print(df['gender'].cat.codes)
```
